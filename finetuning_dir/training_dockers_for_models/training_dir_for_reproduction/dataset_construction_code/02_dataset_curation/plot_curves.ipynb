```python
BASE_FILE = "../../code/00_starter_repo/neurips_llm_efficiency_challenge/sample-submissions/llama_recipes/training_logs"
```

```python
files = list(os.listdir(BASE_FILE))
```

```python
for _file in files:
    _path = os.path.join(BASE_FILE, _file)
    with open(_path, 'r') as fd:
        lines = [x.strip() for x in fd.readlines()]
        lines = list(filter(lambda x:"eval_epoch_loss=tensor(" in x, lines))   
        loss_arr = [float(x.split("eval_epoch_loss=tensor(")[-1].split(",")[0]) for x in lines]
    plt.plot([i+1 for i in range(len(loss_arr))], loss_arr, label = _file)
    #plt.show()
    #break
plt.legend()
```
